# Checking for Outliers

Here I am Using IQR method for to treating the ourliers

# Finding the Outliers in all columns with One Go

```python
for feature in df_numerical:
    data=df.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

# Removing all outliers in single step

```python
# Removing outliers in all columns in one go
for i in df_numerical:
  q1 = np.quantile(df[i],0.25)
  q3 = np.quantile(df[i],0.75)
  iqr = q3-q1
  fence_low  = q1-1.5*iqr
  fence_high = q3+1.5*iqr
  df[i] = np.where(df[i]>fence_high,fence_high,df[i])
  df[i] = np.where(df[i]<fence_low,fence_low,df[i])
```

# **Machine Learning**

**Data Spiliting into training and testing**

```python
# Dividing the dataset into two parts independent and dependent
X = df.drop("Salary",axis=1)
Y = df['Salary']
```

```python
# Spiliting the data into training and testing datasets
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3, random_state=0)
```

**I am taking the best models for checking which one gives the best output**
  
  **1.Random forest**
  
  **2.Decision tree**
  
  **3.Support Vector Regression**
  
  **4.Multiple linear Regression** 

**Model-1: Random Forest**

```python
# Importing random forest and matrices
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_squared_error
randomforest = RandomForestRegressor()
randomforest.fit(X_train,y_train)
```

```python
#predicting the values 
y_predict = randomforest.predict(X_test)
y_predict
```

```python
# checking the performance
MSE = mean_squared_error(y_test,y_predict)
RMSE = np.sqrt(MSE)
r2 = r2_score(y_test,y_predict)
print('The Results is:')
print("RMSE is:{}".format(RMSE))
print("R2 is:{}".format(r2))

# the RMSE value is:0.0020770614690679293
# The R2 Value is:0.9998694330392274
```

# **Output**
